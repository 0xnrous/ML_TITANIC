## Names:
Title: Extracting the title (e.g., Mr, Mrs, Miss) from the passenger name.

```python
# define a function to extract the title from the passenger name
def get_title(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return 'Unknown'

# apply the function to the Name column and create a new column called Title
train['Title'] = train['Name'].apply(get_title)
test['Title'] = train['Name'].apply(get_title)
```

```python
train['Title'].value_counts()
```

```python
test['Title'].value_counts()
```

```python
# combining train and test dataset
train_test_data = [train, test]
```

## Title Mapping : 
we first define a dictionary called **title_map** that maps each title string to its corresponding numerical value. Then, we apply this mapping to the Title column of the dataset using the **map()** method, and create a new column called **Title_Map** to store the mapped values.

```python
train['Title_Map'] = train['Title'].fillna(0)

title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"the Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 , "Unknown": 0 }
for dataset in train_test_data:
    dataset['Title_Map'] = dataset['Title'].map(title_mapping)

```

## Binning: 

**Converting Numerical Age to Categorical Variable**  

feature mapping:  
child: 0  
young: 1  
adult: 2  
mid-age: 3  
senior: 4

```python
# create loop to binning the Age feature into 5 age ranges

import numpy as np

for dataset in train_test_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 62, 'Age'] = 4
    
```

# KNN 

**1.1_Split the data into training and testing sets**

```python
x = train.drop(['Survived'], axis=1)
y = train['Survived']
print (x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split ,GridSearchCV
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7, random_state=1) #split data into 80/20
```

**1.3- train and evaluation of KNN Model**

```python
knn = KNeighborsClassifier(n_neighbors=12)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_train)
accuracy = accuracy_score(y_pred, y_train)
print(round((accuracy)*100,2))
```

**KNN with Cross Validation** 

```python
clf = KNeighborsClassifier(n_neighbors = 12)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# kNN Score
round(np.mean(score)*100, 2)
```

# Hyperparameter tuning for KNN

```python
# Set up the hyperparameter grid
param_grid = {
    'n_neighbors': range(1, 31),
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

# Perform grid search to find the best hyperparameters
grid_search = GridSearchCV(knn, param_grid=param_grid, cv=5)
grid_search.fit(x_train, y_train)
```

```python
# Make the best hyperparameters and score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
# Train the model with the best hyperparameters
knn = KNeighborsClassifier(**best_params)
knn.fit(x_train, y_train)

# Make predictions on the test set and calculate accuracy
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_pred, y_test)
```

```python
print("Best hyperparameters:", best_params)
print("Best cross-validation score:", round(best_score*100,2),'%')
print("Test accuracy:", round(accuracy * 100, 2) ,'%')
```

```python
results_df = pd.DataFrame(grid_search.cv_results_)
pivot_table = results_df.pivot_table(values='mean_test_score', index='param_n_neighbors', columns='param_weights')
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt='.2f')
plt.xlabel('Weights')
plt.ylabel('Number of Neighbors')
plt.title('accuracy scores for KNN with different hyperparameter')
plt.show()
```
